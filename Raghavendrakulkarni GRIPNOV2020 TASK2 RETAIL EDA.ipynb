```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

```python
df=pd.read_csv('SampleSuperstore.csv')
pd.options.display.max_rows = None 
```

```python
df.head()
```

```python
city_sales=df.groupby('City')['Sales'].sum().sort_values(ascending=False)
city_sales
```

### The cities with maximum sales

```python
city_sales[:6].plot(kind='bar')
```

```python
city_profit=df.groupby('City')['Profit'].sum().sort_values(ascending=False)
city_profit
```

### Cities with max profit

```python
city_profit[:6].plot(kind='bar')
```

### Categories from which there are sales

```python
sns.catplot(x="Category",y="Sales",data=df)
```

```python
sns.barplot(x="Sub-Category",y='Sales', data=df)
```

```python
sub_catt=df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
sub_catt.head()
```

### Max sales providing Sub-Categories

```python
sub_catt[:6].plot(kind='bar',color=['red', 'blue', 'purple', 'green', 'lavender','cyan'])
```

### Most common ship mode

```python
plt.figure(figsize=(10,8))
ax = sns.countplot(x="Ship Mode", data=df)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```
